# Rotas otimizadas de bares em Aracaju - SE

    Neste notebook vou exemplificar um projeto que estou desenvolvendo que corresponde a obter a melhor rota. onde pode ser aplicado para empresas de transporte de mercadorias, empresas de turismo, ou até mesmo o alguem que não conhece em detalhes a cidade. A priore estou considerando que o ponto de partida é o mesmo que o ponto de chegada, de modo que é obtido o melhor percurso com relação aos locais que serão visitados.
    
    Neste exemplo escolhi alguns bares localizado em Aracaju-Se.
    
    O endereço de alguns bares foi obtido pelo site guiamais.

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = lambda x: "https://www.guiamais.com.br/encontre?searchbox=true&what=bar&where=Aracaju%2C+SE&page={}".format(x)
address = []
name_bar = []

for i in range(1,6):
    page = requests.get(url(1))

    soup = BeautifulSoup(page.content, 'html.parser')
    blocks = soup.find_all('div', {'class':'free'})

    for block in blocks:
        add = block.find('span')
        if add:
            name_bar.append(block.find('h2',{'class':'aTitle'}).text.strip())
            address.append(add.text.strip())
```

```python
def clear_address(text):
    try:
        text = text.split(',')
        new_text = text[1].strip().split('-')[0].strip()
        new_text += ' '+text[0].strip()
        new_text += ', ' + text[2].split('-')[1].strip()
        new_text += ', ' + text[3].strip()
        return new_text
    except:
        pass
    
address = [clear_address(text) for text in address[:20]] # Vou escolher apenas 20 bares para a analise    
```

----

#### Com os endereços vamos importar a biblioteca folium para plotar o mapa da cidade e a biblioteca que estou desenvolvendo.

```python
import folium
from best_route import Best_Route

routes = Best_Route(address)
```

```python
# Obtendo a latitude e longitude de cada bar #
latlng = routes.dot_location()
med_lat = sum([l['lat'] for l in latlng])/len(latlng) # média latitude
med_lng = sum([l['lng'] for l in latlng])/len(latlng) # média laongitude
```

```python
Map = folium.Map(location=[med_lat, med_lng], zoom_start=12.45, width='60%')

for l in latlng:
    folium.Marker([l['lat'], l['lng']]).add_to(Map)

Map.save('maps/all_bar.html')
Map 
```

    Os bares escolhidos estão um pouco espalhados pela cidade. Vamos supor que você tem 3 dias disponiveis para visitar todos esses bares (rsrs). Deste modo vamos separar esses bares em 3 regiões, para isto vou usar a função KMeans.

----

### Grupos de Bares

    Agora usando a função KMeans vou separar os bares em 3 regiões, de modo que teremos 3 trajetorias minimas.

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([[l['lat'], l['lng']] for l in latlng])

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
# Aqui ilustra a separação dos bares em grupos representados por 0, 1 e 2
kmeans.labels_
```

```python
Map = folium.Map(location=[med_lat, med_lng], zoom_start=12.45, width='60%')

c = ['blue', 'red', 'green']
for i,l in enumerate(X):
    folium.Marker(l, icon=folium.Icon(color=c[kmeans.labels_[i]])).add_to(Map)

Map.save('maps/cluster_bar.html')    
Map 
```

### Definindo a função bar_cluster no qual vai retornar o percurso minimo de carro para cada grupo de bares

```python
def bar_cluster(n=0, _save='maps/melhor_rota_grupo_azul.html'):
    start_end = "929 Avenida Chanceler Osvaldo Aranha, Aracaju, SE"
    address_cluster = [start_end]
    for destiny in np.array(address)[kmeans.labels_==n]:
        if destiny not in address_cluster:
            address_cluster.append(destiny)

    routes_cluster = Best_Route(address_cluster)
    latlng_cluster = routes_cluster.dot_location()
    med_lat_cluster = sum([l['lat'] for l in latlng_cluster])/len(latlng_cluster)
    med_lng_cluster = sum([l['lng'] for l in latlng_cluster])/len(latlng_cluster)
    
    trajeto_min = routes_cluster.min_route(mode='driving')
    
    Map = folium.Map(location=[med_lat_cluster, med_lng_cluster], zoom_start=13, width='60%')
    
    colors = ['red','blue','orange','green','purple','black', 'yellow','darkred',
              'darkblue','darkgreen','lightred','lightgreen',
              'lightblue','darkpurple','cadetblue','lightgray','pink']
    
    for i, l in enumerate(trajeto_min[:,0]):
        if i!=0:        
            folium.Marker([latlng_cluster[l]['lat'], latlng_cluster[l]['lng']],
                          icon=folium.Icon(color=colors[i-1]),
                          popup="Parada {}".format(i)).add_to(Map)
        else:
            folium.Marker([latlng_cluster[l]['lat'], latlng_cluster[l]['lng']],
                          icon=folium.Icon(color='gray'),
                          popup="Casa").add_to(Map)        
        
    directions = routes_cluster.direction_min(mode='driving')
    
    for i, direction in enumerate(directions):
        line = routes_cluster.route(direction)
        folium.PolyLine(line, color=colors[i], weight=2.5, opacity=1).add_to(Map)

    Map.save(_save)
    return Map
```

```python
###  Bares com os Icones Azuis ###
bar_cluster(n=0, _save='maps/melhor_rota_grupo_azul.html')
```
